```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
import os
```

```python
# Initialize executable path for the chromedriver
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
base_url = 'https://www.glassdoor.com/Job/jobs.htm?suggestCount=0&suggestChosen=false&clickSource=searchBtn&typedKeyword=%22data+scientist%22&sc.keyword=%22data+scientist%22&locT=S&locId=3523&jobType='
```

```python
for next_page in range(1,2):
    
    new_url = base_url + "&sc.pageNumber=" + str(next_page)
    browser.visit(new_url)
    html = browser.html
    soup = bs(html, 'html.parser')
    
    tag_card = soup.find('ul', class_="jlGrid")
    
    print(tag_card)
```

```python
 for i in range(12):
        job_title = tag_card.find_all('div', class_='jobTitle')[i].text
        job_location = tag_card.find_all('div', class_='empLoc')[i].text
        job_salary = tag_card.find_all('span', class_='green')[i].text
        #tag_question_ct = tag_card.find_all('span', class_='item-multiplier-count')[i].text
        #tag_week_ct = tag_card.find_all('div', class_='grid--cell stats-row')[i]('a')[1].text
        print(job_title + ' ' + job_location + ' ' + job_salary )    
```
